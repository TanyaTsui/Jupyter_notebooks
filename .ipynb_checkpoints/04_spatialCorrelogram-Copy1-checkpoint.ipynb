```python
af = gpd.read_file('../data/lma/af_keywords.shp')
```

## Correlogram where distance (x axis) is weight distance
example in biology: https://onlinelibrary.wiley.com/doi/pdf/10.1002/cyto.a.23955

```python
def makeNlGrid(cell_size, pc2): # cell_size = 3000
    '''
    cell_size: [float] representing size of cells in meters 
    pc2: [geodataframe] of 2-digit postcodes 
    '''
    # create square grid over NL according to cell size
    xmin, ymin, xmax, ymax= np.array([13565.597251204861, 306846.0765220298, 278026.05770926, 615045.3397637472])
    crs = "EPSG:28992"
    grid_cells = []
    for x0 in np.arange(xmin, xmax+cell_size, cell_size ):
        for y0 in np.arange(ymin, ymax+cell_size, cell_size):
            # bounds
            x1 = x0-cell_size
            y1 = y0+cell_size
            grid_cells.append( shapely.geometry.box(x0, y0, x1, y1)  )
    cell = gpd.GeoDataFrame(grid_cells, columns=['geometry'], crs=crs)

    # cut square grid into NL shape 
    mergeIndex = gpd.sjoin(pc2, cell, how='left', op='intersects').index_right.unique()
    nlGrid = cell.loc[mergeIndex]
    
    return nlGrid

def makeWeightDict(distList, nlGrid): 
    '''
    distList: [list] of distances for weight matrices (in km)
    nlGrid: [gpd.GeoDataFrame] of gridded NL 
    '''
    wDict = {}
    for dist in distList: 
        W = DistanceBand.from_dataframe(df=nlGrid, threshold=dist*1000+100)
        wDict[dist] = W
    return wDict

def makeGrid_cellSize(af, cell_size, mat): 
    
    # turn cell size from km in m
    cell_size = cell_size * 1000 

    # create square grid over NL
    xmin, ymin, xmax, ymax= np.array([13656.74562268, 306988.73757471, 277500.74870041, 612509.05260263])
    crs = "EPSG:28992"
    grid_cells = []
    for x0 in np.arange(xmin, xmax+cell_size, cell_size ):
        for y0 in np.arange(ymin, ymax+cell_size, cell_size):
            # bounds
            x1 = x0-cell_size
            y1 = y0+cell_size
            grid_cells.append( shapely.geometry.box(x0, y0, x1, y1)  )
    cell = gpd.GeoDataFrame(grid_cells, columns=['geometry'], crs=crs)

    # cut square grid into NL shape 
    merged = gpd.sjoin(af, cell, how='left', op='within')
    merged = merged[~merged.index_right.isna()]
    merged.index_right = merged.index_right.map(lambda x: int(x))
    dissolve = merged.dissolve(by="index_right", aggfunc="sum")
    cell.loc[dissolve.index, 'kg'] = dissolve.kg.values
    nlGrid = cell[~cell.kg.isna()].reset_index(drop=True).reset_index()
    nlGrid.rename(columns={'index': 'cellNum'}, inplace=True)

    # put values of material into nlGrid 
    temp = af[af.keyword == mat]
    temp = gpd.sjoin(temp, nlGrid[['cellNum', 'geometry']], how='right')
    temp = temp.dissolve(by="cellNum", aggfunc="sum")[['geometry', 'kg']]
    
    return temp 

def calcMorans(row, af, dataType, distList, log=False, nlGrid=None, wDict=None):
    '''
    this function is applied onto a pd.DataFrame containing a 'mat' column. It is done like this: 
        pd.DataFrame().apply(lambda row: calcMorans(), axis=1)
    
    parameters: 
    row: [pd.Series] containing material name as row.mat 
    af: [gpd.GeoDataFrame] of afgifte dataset with point data, columns = ['keyword', 'kg', 'geometry'], important: no log-transformed values!!
    dataType: [string] indicating datatype, choices = ['point', 'fullGrid', 'partialGrid', 'cellSize']
    nlGrid: [gpd.GeoDataFrame] of NL as a grid 
    wDict: [dictionary] of weight matrices, with distances as keys and matrices as values 
    
    returns: 
    [pd.Series] where index = weight distances, values = morans I values 
    '''
        
    if dataType == 'fullGrid':         
        # create gridded gdf with material
        temp = af[af.keyword == row.mat]
        temp = temp[~temp.geometry.isna()]
        temp = gpd.sjoin(temp, nlGrid, how='right', op='within')
        temp = temp[['kg', 'geometry']].reset_index()
        temp = temp.dissolve(by='index', aggfunc='sum')
        temp = temp.fillna(0)
        if log: 
            temp.kg = temp.kg.map(lambda x: np.log(x) if x > 0 else 0)

        # make list of mis for each weight distance
        mis = []
        np.random.seed(12345) 
        for dist in distList: 
            mi = esda.moran.Moran(temp.kg, wDict[dist]).I
            mis.append(mi)
            
    if dataType == 'partialGrid': 
        # create gridded gdf with material 
        temp = af[af.keyword == row.mat]
        temp = temp[~temp.geometry.isna()]
        temp = gpd.sjoin(temp, nlGrid, how='right', op='within')
        temp = temp[['kg', 'geometry']].reset_index()
        temp = temp.dissolve(by='index', aggfunc='sum')
        temp = temp[~temp.kg.isna()] # remove zero values 
        if log: 
            temp.kg = temp.kg.map(lambda x: np.log(x) if x > 0 else 0)

        # calc morans 
        mis = []
        np.random.seed(12345) 
        for dist in distList: 
            w = DistanceBand.from_dataframe(temp, threshold=dist*1000+100)
            mi = esda.moran.Moran(temp.kg, w).I
            mis.append(mi)

    if dataType == 'point': 
        # create gridded gdf with material 
        temp = af[af.keyword == row.mat]
        temp = temp[~temp.geometry.isna()]
        temp['points'] = temp.geometry.map(lambda x: tuple([x.x, x.y]))
        temp = temp.reset_index()
        if log: 
            temp.kg = temp.kg.map(lambda x: np.log(x) if x > 0 else 0)

        # make list of morans is for each weight distance  
        mis = []
        np.random.seed(12345) 
        for dist in distList: 
            w = DistanceBand(list(temp.points), threshold=dist*1000)
            mi = esda.moran.Moran(temp.kg, w).I
            mis.append(mi)
    
    if dataType == 'cellSize': 
        # for each cell size, create aggregated grid and calculate Moran's I (store values in list)
        mis = []
        np.random.seed(12345)
        for dist in distList: # in this case, dist = cellsize 
            temp = makeGrid_cellSize(af, dist, row.mat) 
            if log: 
                temp.kg = temp.kg.map(lambda x: np.log(x) if x > 0 else 0)
            wq = Queen(temp.geometry)
            mi = esda.moran.Moran(temp.kg, wq).I
            mis.append(mi)
            
    return pd.Series(mis)
```

```python
def makeSpatialCorrelogramSubplots(af, distList, cell_size, log=False): 
    '''
    af: [pd.DataFrame] of afgifte dataset with point geometry, columns = [mat, kg, geometry]
    distList: [list] of distances in km for weight distances. e.g. [3, 5, 10, 20]
    cell_size: [float] indicating size of cell in NL grid, in meters
    log: [boolean] indicating whether data should be log-transformed before moran's I calculation 
    '''
    print('This may take a while, up to 5 minutes, especially if distances in distList are large')
    
    # read / create required data 
    scAll = af.groupby('keyword').sum().reset_index().sort_values(by='kg', ascending=False)
    scAll['nCompanies'] = scAll.keyword.map(lambda x: len(af[af.keyword == x]))
    scAll = scAll[scAll.nCompanies >= 100]
    scAll.rename(columns={'keyword': 'mat'}, inplace=True)
    pc2 = gpd.read_file('../data/spatial-data/nl_pc2_amersfoort.shp')
    
    # create inputs for spatial correlogram 
    nlGrid = makeNlGrid(cell_size, pc2)
    wDict = makeWeightDict(distList, nlGrid)
    
    # make spatial correlogram df for fullGrid, partialGrid, and points 
    scPartialGrid, scFullGrid, scPoint, scCellSize = scAll.copy(), scAll.copy(), scAll.copy(), scAll.copy()
    scPartialGrid[distList] = scPartialGrid.apply(lambda row: calcMorans(row, af, dataType='partialGrid', log=log, distList=distList, nlGrid=nlGrid), axis=1)
    scFullGrid[distList] = scFullGrid.apply(lambda row: calcMorans(row, af, dataType='fullGrid', log=log, distList=distList, nlGrid=nlGrid, wDict=wDict), axis=1)
    scPoint[distList] = scPoint.apply(lambda row: calcMorans(row, af, dataType='point', log=log, distList=distList, nlGrid=nlGrid), axis=1)
    scCellSize[distList] = scCellSize.apply(lambda row: calcMorans(row, af, dataType='cellSize', log=log, distList=distList), axis=1)
        
    # make df for plotting correlogram
    scPointPlot = pd.melt(scPoint, id_vars=['mat'], value_vars=distList)
    scFullGridPlot = pd.melt(scFullGrid, id_vars=['mat'], value_vars=distList)
    scPartialGridPlot = pd.melt(scPartialGrid, id_vars=['mat'], value_vars=distList)
    scCellSizePlot = pd.melt(scCellSize, id_vars=['mat'], value_vars=distList)

    # make dictionary of subplots
    subplotDict = {
        'partial grid': scPartialGridPlot, 
        'full grid': scFullGridPlot, 
        'points': scPointPlot, 
        'cell size': scCellSizePlot
    }
    
    return subplotDict
    
def spatialCorrelogramSubplots(subplotDict, title): 
    # plot correlogram
    from plotly.subplots import make_subplots
    import plotly.graph_objects as go


    fig = make_subplots(rows=1, cols=len(subplotDict), shared_yaxes=True, subplot_titles=tuple(subplotDict.keys()))
    colors = px.colors.qualitative.Plotly * 2
    for i, dfPlot in enumerate(list(subplotDict.values())): 
        for j, mat in enumerate(dfPlot.mat.unique()): 
            temp = dfPlot[dfPlot.mat == mat]
            if i == 0: 
                show = True
            else: 
                show = False
            fig.add_trace(
                go.Scatter(x=temp.variable, y=temp.value, mode='lines+markers', line_color=colors[j], legendgroup=mat, name=mat, showlegend=show),
                row=1, col=i+1
            )
    fig.update_layout(height=600, title_text=title)
    fig.show()
```

## Running functions

```python
distList = [3, 5, 10, 20]
cell_size = 3000
subplotDict_noLog = makeSpatialCorrelogramSubplots(af, distList, cell_size, log=False)
```

## Explanation of charts below
Below are spatial correlograms, where: 
* the vertical y-axis shows the Moran's I value, which indicates how strongly a material is clustered. The higher the value, the stronger the spatial clustering. 
* the horizontal x-axis shows the size of a 'nighborhood' in km when calculating spatial clustering. 
    * For example, 3km means that, when calculating the level of spatial clustering, locations within 3km from each other are considered neighbors, while locations further than 3km from each other are not. 
* the four charts - partial grid, full grid, points, and cell size - are different ways Moran's I is calculated. As you can see, the different methods can have a big difference on the results. 

The correlograms therefore show, for each material, what's the 'ideal' neighborhood size that achieves maximum clustering, which gives some indication of the spatial scale of the material. 
* For example, in the 'points' graph, for asphalt (dark blue line), the line peaks at 5km. This suggests that within clusters of asphalt receivers, companies are within 5km from each other. 
* On the other hand, many other materials (such as wood, fertilizer, food) has the peak value at 3km. This suggests that within clusteres of these materials, companies are within 3km from each other.
